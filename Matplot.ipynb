# customize the line plot with markef and line styles with points

```python
x = np.arange(1,11)
y= np.random.random(10)

fig, ax = plt.subplots(figsize=(16,8))
ax.plot(x,y,marker='o',markerfacecolor='blue',markersize=12,linestyle="--",color="skyblue",linewidth=4)
for i in range(len(x)):
    xi ="{:.1f}".format(x[i])
    yi ="{:.1f}".format(y[i])
    s = str("p("+str(xi)+str(yi)+')')
    ax.text(x[i]+.03,y[i]+.03,s)
    
    
```

# two function in one figure

```python
x=np.linspace(1,10,10)
y1= np.sin(x)
y2= np.cos(x)
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(x,y1,label='sine function')
ax.plot(x,y2,label='Cos function')
ax.set_title('Multiple line plots with legends')
ax.legend()

```

# shareing two axis in one graph


```python
fig,ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time(s)')
ax1.set_ylabel('exponential',color=color)
t=np.arange(1,100)
exp_data =t**2
ax1.plot(t,exp_data,color=color)
ax1.tick_params(axis='y',labelcolor=color)

color ='tab:blue'
ax2 =ax1.twinx()
ax2.set_ylabel('sin funcion',color=color)
sin_data =np.sin(t)
ax2.plot(t,sin_data,color=color)
ax2.tick_params(axis='y',labelcolor=color)
```

```python

fig, ax =plt.subplots()
x=np.linspace(1,20,1000)
ax.plot(x,np.cos(x))
ax.annotate('local maxima',xy=(6.28,1),xytext=(10,5),
            arrowprops=dict(facecolor='black',shrink=.05))
ax.annotate('local minima',xy=(5*np.pi, -1),xytext=(2,-6),
           arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=-90,angleB=0"))
```

```python

import datetime
dateTimeData =[]
date1 = '2011-05-03'
date2 = '2011-05-10'
start = datetime.datetime.strptime(date1, '%Y-%m-%d')
end = datetime.datetime.strptime(date2, '%Y-%m-%d')
step = datetime.timedelta(days=1)
while start <= end:
    dateTimeData.append(str(start.date()))
    start += step
```

```python
x= [int(dateTimeData[date][-2:]) for date in range(len(dateTimeData))]
```

```python
y=np.sin(x)
```

```python
y
```

```python
fig , ax =plt.subplots(figsize=(16,8))
ax.set_xlabel("dates",color="purple")
ax.set_ylabel("sin function",color='green')

plt.xticks(x,dateTimeData)
plt.tick_params(axis='x',colors='Purple',direction='out',length=10,width=5)
ax.plot(x,y,marker='o',markerfacecolor="green",markersize=12,linestyle="--",color="skyblue",linewidth=4)

plt.show()
```
