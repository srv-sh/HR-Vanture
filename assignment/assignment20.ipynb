```python
import pandas as pd
import numpy as np
```

●Read the csv file with the column ‘day’ in date format

```python
df = pd.read_csv("customer_data.csv",parse_dates=['day'])
```

```python
type(df.day[0])
```

●Make the day column the index

```python
df.set_index('day',inplace = True)
```

```python
df
```

●Drop all the rows with blank values in columns - ‘id’,’purchase_status’and ‘name’

```python
new_df=df.dropna(subset=['id','name','purchase_status'])
```

●Drop the rows with all the rows blank

```python
new_df=df.dropna(how='all')
```

```python
new_df
```

●In the columns - age,salary (replace the blank valueswith NaN)

```python
new_df = new_df.replace(
{
    'age':['NaN',' '],
    'salary':['NaN',' ']
},np.nan)
```

●In salary column - remove phrases like ‘dollars’ or‘INR’ using regex

```python
new_df = df.replace({'salary':'[A-Za-z,]*'},'',regex=True)
```

```python
new_df
```

●In the column - age and salary, interpolate the NaNvalues

```python
new_df['salary'] = new_df['salary'].astype('float')
new_df_interpolate = new_df
```

```python

new_df_age = new_df['age'].interpolate()
new_df_salary = new_df['salary'].interpolate()
```

```python
new_df_interpolate['age'] = new_df_age
new_df_interpolate['salary'] = new_df_salary
```

```python
new_df_interpolate
```

●Create the index for missing dates in the dataframe,and fill the missingvalues in a suitable form.

```python
new_df_id = new_df['id'].interpolate(method = 'time')
```

```python
new_df['id']= new_df_id.astype('int')
```

```python
new_df
```

●In the column - age and salary, operate ‘ffill’ and‘bfill’ the NaNvalues(Results has to be saved on a separate dataframe)

```python
new_df_fill = new_df
```

```python

new_df_age = new_df['age'].fillna(method='ffill')
new_df_salary = new_df['salary'].fillna(method='bfill')
new_df_fill['age'] = new_df_age
new_df_fill['salary'] = new_df_salary
```

```python
new_df_fill
```

●In the column ‘Purchase_status’, do a label encoding○0  for ‘not-purchased’○1 for ‘purchased’

```python
new_df_label= df['purchase_status'].fillna(method='ffill')
new_df['purchase_status']=new_df_label
new_df.replace(['not-purchased', 'purchased'], [0,1])
```

Perform one-hot encoding on ‘purchase_status’ column(savethe resultson a separate copy)

```python
new_df_one_hot = pd.get_dummies(new_df,columns=['purchase_status'])
```

```python
new_df_one_hot
```
